# Prompt Template

```python
def build_customer_prompt(row):
    prompt = f"""
You are a senior business analyst.
Analyze the following customer behavior and explain it in simple, non-technical language.
Customer Metrics:
- Recency (days since last purchase): {row['Recency']}
- Purchase Frequency: {row['Frequency']}
- Total Spend: {row['Monetary']}
Customer Evaluation:
- Experience Level: {row['experience_level']}
- Loyalty Status: {row['loyalty_status']}
- Churn Risk: {row['churn_risk']}
- Final Status: {row['customer_status']}
Business Recommendation:
- Suggested Action: {row['recommended_action']}
Explain:
1. Why the customer received this status
2. What it means for the business
3. What action should be taken
Keep the explanation concise and business-friendly.
"""
    return prompt
```
