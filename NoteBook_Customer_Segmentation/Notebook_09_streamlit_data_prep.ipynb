```python
import os
import pandas as pd
```

```python
df = pd.read_csv("processed/customer_status_enriched.csv")
df.head()
```

```python
kpi_summary = {
    "total_customers": len(df),
    "very_happy_pct": round((df['customer_status'] == "Very Happy & Loyal").mean()*100, 2),
    "at_risk_pct": round((df['customer_status'] == "At Risk").mean()*100, 2),
    "churn_risk_pct": round((df['customer_status'] == "Churn Risk").mean()*100, 2),
}
```

```python
revenue_kpi = (
    df.groupby('customer_status')['Monetary']
    .sum()
    .reset_index()
)
```

```python
os.makedirs("app_data", exist_ok=True)
df.to_csv("app_data/customer_data.csv", index=False)
revenue_kpi.to_csv("app_data/revenue_kpi.csv", index=False)
pd.DataFrame([kpi_summary]).to_csv("app_data/kpi_summary.csv", index=False)
```
